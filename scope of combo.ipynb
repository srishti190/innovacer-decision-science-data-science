### Yes there is a scope of bundling dishes into combo. Using one hot encoding technique we get to know which are the items ordered with the corresponding bill number all from the items given in the menu list of Goli's restaurant.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ds2 = pd.read_csv('Customer Order Item Details.csv')
ds2
```

```python
ds2 = pd.read_csv('Customer Order Item Details.csv')
ds2=ds2.rename(columns={"Dish Name":"Dish","Bill Number":"Bill"})
ds2
```

```python
dummies=pd.get_dummies(ds2.Dish)
dummies
```

```python
merged=pd.concat([ds2,dummies],axis='columns')
merged
```

```python
final=merged.drop(['Date','Customer ID','Quantity'],axis='columns')
final.set_index('Bill',inplace=True)
final
```

### here with the corresponding bill number the items which are ordered from the given menu are shown binary 1 and rest by 0

```python
r=final.groupby('Bill')['Dish'].apply(lambda x:"%s" %','.join(x))
p=r.value_counts()
print(p)
#a=np.array(p)
#print(a)
```
