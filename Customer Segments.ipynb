### By what they order

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ds2 = pd.read_csv('Customer Order Item Details.csv')
r=ds2['Bill Number'].value_counts()
r
```

### These are the number of dishes ordered from a particular bill number

```python
a=np.array(r,dtype=int)
print(a)
hm = {}
p = []
q = []
global hm
for i in a: 
    if i in hm: 
        hm[i] += 1
    else: 
        hm[i] = 1
for i in hm:
    
    print("%d %d"%(i,hm[i]))
    p.append(i)
    q.append(hm[i])  
```

```python
x= np.asarray(p)
y= np.asarray(q)
plt.scatter(x,y,label='Data')
plt.xlabel("number of orders")
plt.ylabel("frequency of orders")
plt.title("By what they order")
plt.legend()
plt.show()
```
